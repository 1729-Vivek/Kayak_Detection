```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python

train_dir = 'boats_dataset\Train'
test_dir = 'boats_dataset\TEST'

print("Train set ka structure:")
print(os.listdir(train_dir))
```

```python
cc_train = {}
for cls in os.listdir(train_dir):
    cc_train[cls] = len(os.listdir(os.path.join(train_dir, cls)))


if os.path.exists(test_dir):
    pass
else:
    print("Test directory not found.")

print("Class distribution in the train set:", cc_train)
```

```python
def vis(directory):
    plt.figure(figsize=(10, 10))
    for i, cls in enumerate(os.listdir(directory)):
        img_path = os.path.join(directory, cls, random.choice(os.listdir(os.path.join(directory, cls))))
        img = Image.open(img_path)
        plt.subplot(3, 3, i + 1)
        plt.imshow(img)
        plt.title(cls)
        plt.axis('off')
    plt.show()

print("Visualize karke output:")
vis(train_dir)
```
